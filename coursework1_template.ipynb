```python
sourcelist=np.array(sourcevars)
target_trans=np.array(targetvar)
targetlist=[]
for i in target_trans:
    if i == "tested_positive":
        targetlist.append(1)
    else :
        targetlist.append(0)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
import random
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import roc_curve
from sklearn.metrics import average_precision_score
accuracy_average=0
tpr_average=np.array([0.00,0.00,0.00])
precision_average=0
accuracy_gini=[]
tpr_gini=[]
precision_gini=[]
ran_num=[]
tpr_gini_s=[]

for i in range(1,11):
    random_seed=random.randint(1,1000)
    x_train, x_test, y_train, y_test = train_test_split(sourcelist, targetlist , test_size = 0.3, random_state=random_seed)
    classifier_model = DecisionTreeClassifier(random_state=1234)
    classifier_model = classifier_model.fit(x_train, y_train)
    preds = classifier_model.predict(x_test)
    
    fpr, tpr, third_ = roc_curve(y_test, preds,pos_label=1)
    arr_number=accuracy_score(y_test, preds)
    pre_number=precision_score(y_test, preds,average='macro')
    
    print("The number of the tests is",i)
    print("the random seed is",random_seed)
    print("Accuracy is",arr_number)
    print("True Positive Rate is",tpr)
    print("Precision is",pre_number,"\n")
    
    accuracy_average+=arr_number
    tpr_average+=tpr
    precision_average+=pre_number
    
    accuracy_gini.append(arr_number)
    tpr_gini.append(tpr)
    tpr_gini_s.append(tpr[1])
    precision_gini.append(pre_number)
    ran_num.append(random_seed)
    
print("the average of the Accuracy is",accuracy_average/10,'\n',"the average of the True Positive Rate is",
     0.1*tpr_average,'\n', "the average of the Precision is",precision_average/10)

accuracy_gini.append(accuracy_average/10)

tpr_gini.append(0.1*tpr_average)

precision_gini.append(precision_average/10)

tpr_gini_s.append(0.1*tpr_average[1])
```

```python
ran_str=[str(val) for val in ran_num]
ran_str.append("average result")
result_gini=np.array(list(zip(ran_str,accuracy_gini,tpr_gini,precision_gini)))
pd.DataFrame(result_gini,columns=["random seed ","accuracy","true positive rate","precision"])
```

# Task 2 d

Standardizations(scaling to [0,1] range) would not improve the performance for the dataset, because the standardization would not impact the position of the split point, so, it would not influence the structure of the tree. 
The process of the decision tree is sorting the data by their eigenvalue, and the standardization cannot change eigenvalue and the order of sort, so, the split nodes and leaves would have no difference to the original dataset. 
Moreover, the tree model cannot perform gradient descent because the tree model (regression tree) is built to find the optimal point by finding the optimal split point. Therefore, the tree model is stepped, and the step point is non-directed, and Derivation is meaningless, so normalization is not necessary and have no influence(improvement).

min_sample_split is the minimum number of samples required to split an internal node, which means the data of every point must bigger than the minimum split number to do the next split. Otherwise, progress will be terminated. Every model has the best parameters. Increasing the number of min_sample_split does influence the model, but, we cannot just identify which is the best parameter through the limited iteration.

```python
 x_train, x_test, y_train, y_test = train_test_split(sourcelist, targetlist , test_size = 0.3, random_state=123)
```

```python
split_number=[2,5,10,15]
ah=[]
for each in split_number :
    classifier_model = DecisionTreeClassifier(min_samples_split= each,random_state=1234)
    classifier_model =  classifier_model.fit(x_train, y_train)
    preds =  classifier_model.predict(x_test)
    ah.append(accuracy_score(y_test, preds))
    print("min_sample_split is",each,"Accuracy is",accuracy_score(y_test, preds))
```

```python
bh=['2','5','10','15']
c = np.array(list(zip(bh,ah)))

pd.DataFrame(c,
                   columns=["split number","accuray"])

```

```python
from prettytable import PrettyTable


table=PrettyTable()
table.add_column("Order number",[1,2,3,4])
table.add_column("Min sample slite number",bh)
table.add_column("Accuracy",ah)

print(table)
```

max_depth limit the max depth of the tree. Every note's depth number bigger than it needs to be cut. If None, then nodes are expanded until all leaves are pure or until all leaves contain less than min_samples_split samples.  Increasing the number of max_depth does influence the model, but, we cannot just identify which is the best parameter through the limited iteration.

```python
depth_number=[3,4,5,6]
acc=[]
for f in depth_number:
    classifier_model = DecisionTreeClassifier(max_depth=f,random_state=1234)
    classifier_model =  classifier_model.fit(x_train, y_train)
    preds =  classifier_model.predict(x_test)
    acc.append(accuracy_score(y_test, preds))
    print("Max_ depth is",f,"Accuracy is",accuracy_score(y_test, preds))
```

```python
de_n=['3','4','5','6']
df=np.array(list(zip(de_n,acc)))
pd.DataFrame(df,
                   columns=["depth_number","accuray"])
```

```python
from prettytable import PrettyTable


table=PrettyTable()
table.add_column("Order number",[1,2,3,4])
table.add_column("Max Depth Number",de_n)
table.add_column("Accuracy",acc)

print(table)
```
